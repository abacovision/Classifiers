```python
import pandas as pd
#Downgraded numpy for compatibility purposes
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle
```

```python
def unpack_pickle(path_train):
    try:
        if type(path_train) != str:
            raise ValueError
        p = Path(path_train)
        #Read file for training and testing data.
        if p.exists():
            with open(p, 'rb') as train:
                df = pickle.load(train, encoding='bytes')
            return df
        else:
            raise FileNotFoundError
    except Exception as e:
        raise e
```

```python
batch_1 = unpack_pickle('dataset/data_batch_1')
batch_2 = unpack_pickle('dataset/data_batch_2')
batch_3 = unpack_pickle('dataset/data_batch_3')
batch_4 = unpack_pickle('dataset/data_batch_4')
batch_5 = unpack_pickle('dataset/data_batch_5')
```
