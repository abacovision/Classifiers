```python
import pandas as pd
#Downgraded numpy for compatibility purposes
import numpy as np
import matplotlib.pyplot as plt
import pickle
from pathlib import Path
from scipy.io import loadmat
import os
```

```python
def unpack_df(path_train, path_test):
    try:
        if type(path_train) != str or type(path_test) != str:
            raise TypeError
        p = Path(path_train)
        q = Path(path_test)
        #Read file for training and testing data.
        if p.exists() and q.exists():
            os.path.abspath(os.path.dirname(__file__))
            with open(p, 'rb') as train, open(q, 'rb') as test:
                train_df = loadmat(train.read(), spmatrix=False)
                test_df = loadmat(train.read(), spmatrix=False)
                train.close()
                test.close()
                
            return train_df, test_df
        else:
            raise FileNotFoundError
    except Exception as e:
        print(e)
```
