```python
import pandas as pd
#Downgraded numpy for compatibility purposes
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle
```

```python
def unpack_pickle(path_train):
    try:
        if type(path_train) != str:
            raise ValueError
        p = Path(path_train)

        if p.exists():
            with open(p, 'rb') as train:
                df_temp = pickle.load(train, encoding='bytes')
                X = df_temp[b'data']
                Y = df_temp[b'labels']
                X = X.reshape(10000, 32, 32, 3)
                
            return X, Y
        else:
            raise FileNotFoundError
    
    except Exception as e:
        raise e
```
