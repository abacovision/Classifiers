```python
import pandas as pd
#Downgraded numpy for compatibility purposes
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle
```

```python
def train(path_train):
    try:
        if type(path_train) != str:
            raise ValueError


        p = Path(path_train)
        rnd = np.random.default_rng()
        bias = rnd.random(3072,)
        bias = np.transpose(bias)
        if p.exists():
            #OPEN THE FILE
            with open(p, 'rb') as train:
                #UNPICKLE THE FILE
                df_temp = pickle.load(train, encoding='bytes')
                #TAKE JUST WHAT WE NEED
                X = df_temp[b'data']
                Y = np.array(df_temp[b'labels'])
                #RESHAPE FEATURES
                X = X.reshape(10000,3072)
            #CREATE WEIGHTS VECTOR
            weights = rnd.random((10,3072))
            

            return X, Y
        else:
            raise FileNotFoundError(path_train)
        return weights, X, Y, bias
    except Exception as e:
        raise e
```

```python
X, Y = train('dataset/data_batch_1')
```
