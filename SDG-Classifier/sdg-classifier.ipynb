```python
#Importing dataset, model and the plotter framework
from sklearn import datasets
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
```

```python
#Defining the model and import the dataset
iris = datasets.load_iris()
model = SGDClassifier()
```

```python
#Plotting the data
fig, ax = plt.subplots()
scatter = plt.scatter(iris.data[:,0], iris.data[:,1], c=iris.target)
ax.set(ylabel='Sepal length', xlabel='Sepal width')
fig = ax.legend(scatter.legend_elements()[0], ['Setosa', 'Versicolor', 'Virginica'], loc="lower right", title="Classes")
```
