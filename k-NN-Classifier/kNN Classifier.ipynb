```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
```

```python
with open('heart.csv','r') as f:
    df = pd.read_csv(f)
    f.close()
```

```python
def preprocess(df_, header_list, column_drop):
    try:
    
        dataframe = df_
        if column_drop != None:
            dataframe = dataframe.drop(column_drop, axis=1)
        
        p = re.compile('[a-z0-9]', re.IGNORECASE)
        rgx = lambda x: "".join(p.findall(x))
        lowercase = lambda x: x.lower()
        #Hash to parse str object to int.
        converting = lambda x: hash(x)
        
        for i in header_list:
            dataframe[i] = dataframe[i].apply(rgx)
            dataframe[i] = dataframe[i].apply(lowercase)
            dataframe[i] = dataframe[i].apply(converting)
        return dataframe
    
    except Exception as e:
        raise e

def slice_data(x):
    try:
        shape = 800
        test_data = x.loc[shape:,]
        return test_data
    except Exception as e:
        raise e
```

```python
test_data = slice_data(df)
df = preprocess(df, ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], 'HeartDisease')
test_data = preprocess(test_data, ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], None)
```
